```python
import pandas as pd
import numpy as np
from scipy.stats import stats
```

```python
data=pd.read_excel('C:\\Users\\DELL\\Desktop\\ATI\\python\\proj_data.xlsx')
```

```python
data.columns
```

```python
#Chi square test of independence is used for hypothesis testing since 
#there are categoric variables only(categoric variables have only frequency/count/mode)

from scipy import stats
#Hypothesis 1: 
#Ho:Training delivery has no significant impact on employeeperformance.
#Ha: Training delivery hassignificant impact on employee performance.
crosstab1 = pd.crosstab(data['Training_Delivery'], data['Emploee_Performance'])
crosstab1
stats.chi2_contingency(crosstab1)
```

```python
#Hypothesis 2:
#Ho:Training transfer has no significant impact on employeeperformance.
#Ha: Training transfer has significant impact on employee performance.
crosstab2 = pd.crosstab(data['Training_Transfer'], data['Emploee_Performance'])
crosstab2
stats.chi2_contingency(crosstab2)
```

```python
#Hypothesis 3:
#Ho: Motivation has no significant impact on employee performance.
#Ha: Motivation has significant impact on employee performance.
crosstab3 = pd.crosstab(data['Employee_Motivation'], data['Emploee_Performance'])
crosstab3
stats.chi2_contingency(crosstab3)
```

```python
#Hypothesis 4:
#Ho:Training delivery has no significant impact on training transfer.
#Ha: Training delivery has significant impact on training transfer.
crosstab4 = pd.crosstab(data['Training_Delivery'], data['Training_Transfer'])
crosstab4
stats.chi2_contingency(crosstab4)
```

```python
#Hypothesis 5:
#Ho:Training delivery has no significant impact on employee motivation.
#Ha: Training delivery has significant impact on employee motivation.
crosstab5 = pd.crosstab(data['Training_Delivery'], data['Employee_Motivation'])
crosstab5
stats.chi2_contingency(crosstab5)
```

```python
#Hypothesis 6:
#Ho:Employee motivation has no significant impact on training transfer.
#Ha:Employee motivation has significant impact on training transfer.
crosstab6 = pd.crosstab(data['Employee_Motivation'], data['Training_Transfer'])
crosstab6
stats.chi2_contingency(crosstab6)
```

Output:
The first value is the Chi-square value, followed by the p-value, then comes the degrees of freedom (3), and lastly it outputs the expected frequencies as an array. If all of the expected frequencies are greater than 5, the chi2 test results can be trusted. We can reject the null hypothesis if the p-value is less than 0.05

From the above test results all the p values are <0.05, thus the null hypotheses are rejected.
However, the results of the chi square tests cannot be trusted since the expected frequencies are not >5.
